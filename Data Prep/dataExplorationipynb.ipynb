```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
aaplDf = pd.read_csv('/content/aaplNewsStock.csv')
msftDf = pd.read_csv('/content/msftNewsStock.csv')
djiaDf = pd.read_csv('/content/djiaNewsStock.csv')
```

```
aaplDf = aaplDf.drop(columns = ['Adj Close', 'Close' , 'High' , 'Low', 'Volume','News'])
msftDf = msftDf.drop(columns = ['Adj Close', 'Close' , 'High' , 'Low', 'Volume','News'])
djiaDf = djiaDf.drop(columns = ['Adj Close', 'Close' , 'High' , 'Low', 'Volume','News'])
```

## Plot Histograms to Show Near Even Spread of the Market Going Up and Down

### Apple

```
sns.distplot(aaplDf['Label'],label = 'AAPL')
```

### Microsoft

```
sns.distplot(msftDf['Label'])
```

### Dow Jones

```
sns.distplot(djiaDf['Label'])
```

## Plot Histograms to Visualize the Distribution of Sentiments

### Apple

```
sns.distplot(aaplDf['compound'],label = 'AAPL')
```

### Microsoft

```
sns.distplot(msftDf['compound'], label = 'MSFT')
```

### Dow Jones

```
sns.distplot(djiaDf['compound'], label = 'DJIA')
```

### Apple and Microsft Sentiment Comparison

```
sns.distplot(aaplDf['compound'],label = 'AAPL')
sns.distplot(msftDf['compound'], label = 'MSFT')
```
